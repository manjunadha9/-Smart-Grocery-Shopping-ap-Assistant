```python
#try to run this code before runing above code 
import pandas as pd

def encode_units(x):
  try:
    if x < 1:
      return 0
    if x >= 1:
      return 1
  except TypeError:
    return None  


basket = pd.to_numeric(basket, errors='coerce')  


basket_encoded = (basket >= 1).astype(int)


print(basket_encoded.head(10))
```

# observations in category 

```python
plt.figure(figsize=(15,3))
plt.subplot(1,2,1)
sns.countplot(x='category', data=df)
plt.subplot(1,2,2)
sns.countplot(x='product_name', data=df)
plt.show()
```

```python
weekday_sales = df.groupby('day_of_week')['total_revenue'].sum().sort_values(ascending=False)
best_weekday_sales = weekday_sales.head(10)
plt.figure(figsize=(10, 6))
best_weekday_sales.sort_values().plot(kind='barh', color=['lightgreen'])
plt.title('Top Selling weekdays by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('weekdays ')
plt.grid(axis='x')
plt.show()
```

total revenue generated on holidays versus regular days

```python
df['is_holiday'] = df['holiday'].map({True: 'Holiday', False: 'Regular Day'})
revenue_by_holiday = df.groupby('is_holiday')['total_revenue'].sum()
plt.figure(figsize=(8, 6))
revenue_by_holiday.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Total Revenue Comparison: Holidays vs. Regular Days')
plt.xlabel('Day Type')
plt.ylabel('Total Revenue')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

top-selling products

```python
top_selling_products = df.groupby('product_name')['total_revenue'].sum().sort_values(ascending=False)
top_10_products = top_selling_products.head(10)
plt.figure(figsize=(10, 6))
top_10_products.sort_values().plot(kind='barh', color=['skyblue'])
plt.title('Top Selling Products by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')
plt.grid(axis='x')
plt.show()
```

gender distribution of buyers

```python
gender_distribution = df['buyer_gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', colors=['lightpink', 'lightblue'], startangle=140)
plt.title('Gender Distribution of Buyers')
plt.axis('equal') 
plt.show()
```

```python
df['sales_date'] = pd.to_datetime(df['sales_date'])
daily_revenue = df.groupby('sales_date')['total_revenue'].sum()
plt.figure(figsize=(10, 6))
plt.plot(daily_revenue.index, daily_revenue.values, color='cadetblue', marker='o', linestyle='-')
plt.title('Daily Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
df['month'] = df['sales_date'].dt.month
df['quarter'] = df['sales_date'].dt.quarter
monthly_revenue = df.groupby('month')['total_revenue'].sum()
quarterly_revenue = df.groupby('quarter')['total_revenue'].sum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', linestyle='-')
plt.title('Total Revenue Trend Over Different Months')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['price'], df['number_of_items_sold'], alpha=0.7)
plt.title('Relationship between Price and Number of Items Sold')
plt.xlabel('Price')
plt.ylabel('Number of Items Sold')
plt.grid(True)
plt.show()
```
